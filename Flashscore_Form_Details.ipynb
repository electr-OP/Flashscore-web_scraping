```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html)
```

```python
results = soup.find_all('div', attrs={'title' : 'Click for match detail!'})
```

```python
results
```

```python
home_t = []
away_t = []
for result in results:
    home_t.append(result.find('div', attrs = {'event__participant--home'}).text)
    away_t.append(result.find('div', attrs = {'event__participant--away'}).text)

```

```python
h_odd = []
a_odd = []

for result in results:
    odds_h = result.find('div', attrs={'class' : 'o_1'})
    odds_a = result.find('div', attrs={'class' : 'o_2'})
    
    if odds_h == None or odds_a == None:
        h_odd.append('-')
        a_odd.append('-')
    
    else:
    
        #odd_h = odds_h.text

        #odd_a = odds_a.text
        
        h_odd.append(odds_h.text)
        a_odd.append(odds_a.text)
    
```

```python
Sh_odd = pd.Series(h_odd)
```

```python
Sa_odd = pd.Series(a_odd)
```

```python
Sh_odd = pd.to_numeric(Sh_odd, errors = 'coerce')
Sa_odd = pd.to_numeric(Sa_odd, errors = 'coerce')
```

```python
while k < len(Sh_odd):
    print(abs(Sh_odd[k] - Sa_odd[k]))
    k+=1
```

```python
link_list = []
k = 0
while k < len(results):
    if abs(Sh_odd[k] - Sa_odd[k]) >= 4:
        link_id = results[k].get_attribute_list('id')[0][4:]
        link_list.append(link_id)
    k+=1
    
len(link_list)
```

```python
flink_listh2h = []
flink_liststand = []
for L in link_list:
    FLh = 'https://www.flashscore.com.ng/match/'+ L + '#h2h;overall'
    FLs = 'https://www.flashscore.com.ng/match/'+ L + '#standings;table;overall'
    flink_listh2h.append(FLh)
    flink_liststand.append(FLs)
print(flink_listh2h)
```

```python
def flash_statsstand(link):
    
    
    driver.get(link)
    
    html = driver.page_source

    soup = BeautifulSoup(html)
    
    table = []
    
    
    high_teams = soup.find_all('div', attrs={'class' : 'highlight'})
    
    if high_teams == []:
        home_tp = '-'
        away_tp = '-'
        lg_fmhW = '-'
        lg_fmhD = '-'
        lg_fmhL = '-'
        lg_fmaW = '-'
        lg_fmaD = '-'
        lg_fmaL = '-'
        
    
    for t in high_teams:
        for n in home_t:
            if t.find('span', attrs = {'class' : 'team_name_span'}).text == n:
                home_tp = t.find('div', attrs={'class' : 'cell--sorted'}).text[3:-3]
        for n in away_t:
            if t.find('span', attrs = {'class' : 'team_name_span'}).text == n:
                away_tp = t.find('div', attrs={'class' : 'cell--sorted'}).text[3:-3]
                
    lg_fmlH = []
    lg_fmlA = []
    for t in high_teams:
        for n in home_t:
            if t.find('span', attrs = {'class' : 'team_name_span'}).text == n:
                for p in t.find_all('a', attrs={'class' : 'form__cell'})[1:6]:
                    lg_fmlH.append(p.text)

        for n in away_t:
            if t.find('span', attrs = {'class' : 'team_name_span'}).text == n:
                for p in t.find_all('a', attrs={'class' : 'form__cell'})[1:6]:
                    lg_fmlA.append(p.text)
                    
    lg_fmhW = lg_fmlH[0:5].count('W')

    lg_fmhD = lg_fmlH[0:5].count('D')

    lg_fmhL = lg_fmlH[0:5].count('L')
    
    lg_fmaW = lg_fmlA[0:5].count('W')

    lg_fmaD = lg_fmlA[0:5].count('D')

    lg_fmaL = lg_fmlA[0:5].count('L')
                            
    table.append((home_tp, away_tp, lg_fmhW, lg_fmhD, lg_fmhL, lg_fmaW, lg_fmaD, lg_fmaL))
    import pandas as pd
    table = pd.DataFrame(table, columns= ['home_tp', 'away_tp', 'lg_fmhW', 'lg_fmhD', 'lg_fmhL', 'lg_fmaW', 'lg_fmaD', 'lg_fmaL'])
    return table
   
```

```python
def flash_statsh2h(link):
    
    driver.get(link)
    
    html = driver.page_source

    soup = BeautifulSoup(html)
    
    table = [] 
    
    home_t = soup.find('div', attrs={'class' : 'tname-home'}).text[5:-3]
    
    away_t = soup.find('div', attrs={'class' : 'tname-away'}).text[3:-5]
    
    odds_h = soup.find('td', attrs={'class' : 'o_1'})
    odds_a = soup.find('td', attrs={'class' : 'o_2'})
    
    if odds_h == None or odds_a == None:
        odd_h = '-'
        odd_a = '-'
    else:
    
        odd_h = odds_h.text[4:-2]

        odd_a = odds_a.text[4:-2]
    
    ov_h2home = soup.find('table', attrs = {'class' : 'h2h_home'})
    
    ov_h2away = soup.find('table', attrs = {'class' : 'h2h_away'})
    
    h2h_home = soup.find('div', attrs = {'id' : 'tab-h2h-home'})
    
    h2h_away = soup.find('div', attrs = {'id' : 'tab-h2h-away'})

    ov_h2homel = []
    for t in ov_h2home.find_all('a', attrs = {'class' : 'wld'})[0:5]:
        ov_h2homel.append(t.text)
        
    ov_h2homeW = ov_h2homel.count('W')

    ov_h2homeD = ov_h2homel.count('D')

    ov_h2homeL = ov_h2homel.count('L')
    
    h2h_homel = []
    for t in h2h_home.find_all('a', attrs = {'class' : 'wld'})[0:5]:
        h2h_homel.append(t.text)
        
    h2h_homeW = h2h_homel.count('W')

    h2h_homeD = h2h_homel.count('D')

    h2h_homeL = h2h_homel.count('L')
    
    h2h_awayl = []
    for t in h2h_away.find_all('a', attrs = {'class' : 'wld'})[0:5]:
        h2h_awayl.append(t.text)
        
    h2h_awayW = h2h_awayl.count('W')

    h2h_awayD = h2h_awayl.count('D')

    h2h_awayL = h2h_awayl.count('L')
    
    table.append((home_t, away_t, odd_h, odd_a, ov_h2homeW, ov_h2homeD, ov_h2homeL, h2h_homeW, h2h_homeD, h2h_homeL, h2h_awayW, h2h_awayD, h2h_awayL))
    
    import pandas as pd
    table = pd.DataFrame(table, columns= ['home_t', 'away_t', 'odd_h', 'odd_a', 'ov_h2homeW', 'ov_h2homeD', 'ov_h2homeL', 'h2h_homeW', 'h2h_homeD', 'h2h_homeL', 'h2h_awayW', 'h2h_awayD', 'h2h_awayL'])

    return table
```
